```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# ‚úÖ Set your filename (no extension)
file_name = 'bitcoin2018_LSTM_tanh_relu'

# ‚úÖ Load the result CSV
csv_path = os.path.join('result', file_name + '.csv')
df = pd.read_csv(csv_path)

# ‚úÖ Optional: preview columns
print("üîç CSV Columns:", df.columns.tolist())

# ‚úÖ Plotting loss values if available
plt.figure(figsize=(20, 10))

# If your CSV has `epoch`, `loss`, and `val_loss`
if 'val_loss' in df.columns:
    sns.boxplot(data=df[['loss', 'val_loss']])
    plt.title(f"üìä Box Plot of Training and Validation Loss - {file_name}")
else:
    # If the CSV has multiple columns for different runs (like different L1/L2 configs)
    df = df.set_index(df.columns[0])  # set epoch or unnamed column as index
    df.boxplot()
    plt.title(f"üìä Box Plot of Multiple LSTM Configurations - {file_name}")

plt.yscale('log')
plt.ylabel('Loss (log scale)')
plt.grid(True)

# ‚úÖ Save and show
plot_path = os.path.join('result', file_name + '_boxplot.png')
plt.savefig(plot_path)
plt.show()

print(f"‚úÖ Box plot saved to: {plot_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# === [üìÅ Config] ===
file_name = 'bitcoin2018_LSTM_tanh_relu'  # change if you used a different name
csv_path = os.path.join('result', file_name + '.csv')

# === [üìÑ Load CSV] ===
df = pd.read_csv(csv_path)

# === [üîç Show Columns] ===
print("Columns in CSV:", df.columns.tolist())

# === [üìä Boxplot of Loss vs Val_Loss] ===
plt.figure(figsize=(16, 8))
sns.boxplot(data=df[['loss', 'val_loss']])
plt.title('Boxplot of Training and Validation Loss')
plt.ylabel('MSE Loss')
plt.yscale('log')  # set to log scale for better visibility
plt.grid(True)

# === [üíæ Save Plot] ===
plot_path = os.path.join('result', file_name + '_boxplot.png')
plt.savefig(plot_path)
plt.show()

print(f"‚úÖ Saved boxplot to: {plot_path}")
```
